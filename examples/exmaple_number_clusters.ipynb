```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage

def l_method(eval_graph, min_clusters=2, max_clusters=None):
    """
    使用L方法确定最佳聚类数量
    
    参数:
    eval_graph: 形状为(n, 2)的numpy数组，包含(聚类数量, 评估指标)对
    min_clusters: 最小考虑的聚类数量
    max_clusters: 最大考虑的聚类数量
    
    返回:
    optimal_k: 最佳聚类数量
    """
    # 确保数据按聚类数排序
    data = np.array(eval_graph)
    data = data[data[:, 0].argsort()]
    
    # 应用聚类数限制
    if min_clusters > 2:
        data = data[data[:, 0] >= min_clusters]
    if max_clusters is not None:
        data = data[data[:, 0] <= max_clusters]
    
    n = len(data)
    if n <= 3:
        return int(data[n//2, 0])
    
    min_rmse = float('inf')
    knee_idx = 0
    
    # 遍历所有可能的拐点位置
    for c in range(2, n-2):
        # 左侧序列 Lc
        Lc = data[:c+1]
        # 右侧序列 Rc
        Rc = data[c+1:]
        
        # 左侧线性回归
        X_left = Lc[:, 0].reshape(-1, 1)
        y_left = Lc[:, 1]
        model_left = LinearRegression().fit(X_left, y_left)
        y_pred_left = model_left.predict(X_left)
        rmse_left = np.sqrt(np.mean((y_left - y_pred_left) ** 2))
        
        # 右侧线性回归
        X_right = Rc[:, 0].reshape(-1, 1)
        y_right = Rc[:, 1]
        model_right = LinearRegression().fit(X_right, y_right)
        y_pred_right = model_right.predict(X_right)
        rmse_right = np.sqrt(np.mean((y_right - y_pred_right) ** 2))
        
        # 计算加权总RMSE
        w_left = len(Lc) / float(n)
        w_right = len(Rc) / float(n)
        total_rmse = w_left * rmse_left + w_right * rmse_right
        
        if total_rmse < min_rmse:
            min_rmse = total_rmse
            knee_idx = c
    
    # 返回拐点对应的聚类数量
    return int(data[knee_idx, 0])

def iterative_l_method(eval_graph, min_clusters=2):
    """
    使用迭代改进版的L方法确定最佳聚类数量
    
    参数:
    eval_graph: 评估图数据
    min_clusters: 最小考虑的聚类数量
    
    返回:
    optimal_k: 最佳聚类数量
    """
    data = np.array(eval_graph)
    data = data[data[:, 0].argsort()]
    data = data[data[:, 0] >= min_clusters]
    
    # 初始化
    last_knee = len(data)
    current_knee_idx = len(data) - 1
    cutoff = None
    
    # 迭代直到收敛
    while True:
        # 如果没有指定cutoff，则使用全部数据
        if cutoff is None:
            focus_data = data
        else:
            # 只关注cutoff以内的数据点
            focus_data = data[data[:, 0] <= cutoff]
            
            # 确保我们有足够的数据点
            if len(focus_data) < 5:  # 确保至少有5个点
                focus_data = data[:min(5, len(data))]
        
        # 找到拐点
        n = len(focus_data)
        min_rmse = float('inf')
        knee_idx = 0
        
        for c in range(1, n-1):
            # 左侧和右侧序列
            Lc = focus_data[:c+1]
            Rc = focus_data[c+1:]
            
            if len(Lc) < 2 or len(Rc) < 2:
                continue
            
            # 左侧线性回归
            X_left = Lc[:, 0].reshape(-1, 1)
            y_left = Lc[:, 1]
            model_left = LinearRegression().fit(X_left, y_left)
            y_pred_left = model_left.predict(X_left)
            rmse_left = np.sqrt(np.mean((y_left - y_pred_left) ** 2))
            
            # 右侧线性回归
            X_right = Rc[:, 0].reshape(-1, 1)
            y_right = Rc[:, 1]
            model_right = LinearRegression().fit(X_right, y_right)
            y_pred_right = model_right.predict(X_right)
            rmse_right = np.sqrt(np.mean((y_right - y_pred_right) ** 2))
            
            # 计算加权总RMSE
            w_left = len(Lc) / float(n)
            w_right = len(Rc) / float(n)
            total_rmse = w_left * rmse_left + w_right * rmse_right
            
            if total_rmse < min_rmse:
                min_rmse = total_rmse
                knee_idx = c
        
        current_knee = int(focus_data[knee_idx, 0])
        
        # 如果拐点不再向左移动或者达到最小聚类数，则收敛
        if current_knee >= last_knee or current_knee <= min_clusters:
            break
        
        # 更新结果，并设置新的cutoff（拐点的2倍）
        last_knee = current_knee
        cutoff = current_knee * 2
    
    return current_knee

def visualize_l_method(eval_graph, optimal_k, title="L方法确定最佳聚类数"):
    """
    可视化L方法的结果
    
    参数:
    eval_graph: 评估图数据
    optimal_k: L方法确定的最佳聚类数量
    """
    data = np.array(eval_graph)
    data = data[data[:, 0].argsort()]
    
    x = data[:, 0]
    y = data[:, 1]
    
    # 找到最佳k对应的索引
    k_idx = np.where(x == optimal_k)[0][0]
    
    # 左右分割数据
    left_data = data[:k_idx+1]
    right_data = data[k_idx:]
    
    # 左侧线性回归
    X_left = left_data[:, 0].reshape(-1, 1)
    y_left = left_data[:, 1]
    model_left = LinearRegression().fit(X_left, y_left)
    y_pred_left = model_left.predict(X_left)
    
    # 右侧线性回归
    X_right = right_data[:, 0].reshape(-1, 1)
    y_right = right_data[:, 1]
    model_right = LinearRegression().fit(X_right, y_right)
    y_pred_right = model_right.predict(X_right)
    
    # 绘图
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, c='blue', label='评估图数据点')
    plt.plot(X_left.flatten(), y_pred_left, 'r-', linewidth=2, label='左侧拟合线')
    plt.plot(X_right.flatten(), y_pred_right, 'g-', linewidth=2, label='右侧拟合线')
    plt.axvline(x=optimal_k, color='purple', linestyle='--', label=f'最佳聚类数 (k={optimal_k})')
    
    plt.title(title)
    plt.xlabel('聚类数量')
    plt.ylabel('评估指标值')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    
    return plt

class AutoCluster:
    """
    使用L方法自动确定最佳聚类数量的聚类算法
    """
    def __init__(self, min_clusters=2, max_clusters=20, method='hierarchical', 
                 linkage='ward', metric='euclidean', affinity='euclidean',
                 eval_metric='distortion', l_method_type='iterative'):
        """
        初始化
        
        参数:
        min_clusters: 最小考虑的聚类数量
        max_clusters: 最大考虑的聚类数量
        method: 聚类算法 ('hierarchical', 'kmeans')
        linkage: 层次聚类的链接方法 ('ward', 'complete', 'average', 'single')
        metric: 距离度量
        affinity: 层次聚类的亲和度度量
        eval_metric: 用于L方法的评估指标 ('distortion', 'silhouette', 'calinski_harabasz', 'davies_bouldin')
        l_method_type: L方法类型 ('standard', 'iterative')
        """
        self.min_clusters = min_clusters
        self.max_clusters = max_clusters
        self.method = method
        self.linkage = linkage
        self.metric = metric
        self.affinity = affinity
        self.eval_metric = eval_metric
        self.l_method_type = l_method_type
        
        self.optimal_k = None
        self.labels_ = None
        self.eval_results = None
        self.cluster_centers_ = None
    
    def fit(self, X):
        """
        拟合聚类模型并自动确定最佳聚类数量
        
        参数:
        X: 输入数据，形状为(n_samples, n_features)
        
        返回:
        self
        """
        # 标准化数据
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
        
        # 计算不同聚类数量下的评估指标
        k_values = range(self.min_clusters, self.max_clusters + 1)
        eval_values = []
        
        for k in k_values:
            # 应用聚类算法
            if self.method == 'hierarchical':
                model = AgglomerativeClustering(
                    n_clusters=k, 
                    linkage=self.linkage,
                    metric=self.affinity
                )
                labels = model.fit_predict(X_scaled)
            elif self.method == 'kmeans':
                model = KMeans(n_clusters=k, random_state=42)
                labels = model.fit_predict(X_scaled)
                if k == self.max_clusters:
                    self.cluster_centers_ = model.cluster_centers_
            else:
                raise ValueError(f"不支持的聚类方法: {self.method}")
            
            # 计算评估指标
            if self.eval_metric == 'distortion':
                if self.method == 'kmeans':
                    # K-means的惯性
                    score = model.inertia_
                else:
                    # 计算簇内距离的平方和
                    score = 0
                    for cluster_id in range(k):
                        cluster_points = X_scaled[labels == cluster_id]
                        if len(cluster_points) > 0:
                            # 计算到质心的距离平方和
                            centroid = np.mean(cluster_points, axis=0)
                            score += np.sum((cluster_points - centroid) ** 2)
            elif self.eval_metric == 'silhouette':
                if len(np.unique(labels)) > 1:  # 至少需要2个簇
                    score = -silhouette_score(X_scaled, labels)  # 负值使得评估图形状与其他指标一致
                else:
                    score = -0.1  # 占位值
            elif self.eval_metric == 'calinski_harabasz':
                if len(np.unique(labels)) > 1:  # 至少需要2个簇
                    score = -calinski_harabasz_score(X_scaled, labels)  # 负值
                else:
                    score = -1  # 占位值
            elif self.eval_metric == 'davies_bouldin':
                if len(np.unique(labels)) > 1:  # 至少需要2个簇
                    score = davies_bouldin_score(X_scaled, labels)
                else:
                    score = 10  # 占位值
            else:
                raise ValueError(f"不支持的评估指标: {self.eval_metric}")
            
            eval_values.append(score)
        
        # 创建评估图
        self.eval_results = np.column_stack((k_values, eval_values))
        
        # 应用L方法确定最佳聚类数量
        if self.l_method_type == 'standard':
            self.optimal_k = l_method(self.eval_results, 
                                      min_clusters=self.min_clusters, 
                                      max_clusters=self.max_clusters)
        else:  # 'iterative'
            self.optimal_k = iterative_l_method(self.eval_results, 
                                               min_clusters=self.min_clusters)
        
        # 使用最佳聚类数量重新拟合模型
        if self.method == 'hierarchical':
            model = AgglomerativeClustering(
                n_clusters=self.optimal_k, 
                linkage=self.linkage,
                metric=self.affinity
            )
            self.labels_ = model.fit_predict(X_scaled)
        elif self.method == 'kmeans':
            model = KMeans(n_clusters=self.optimal_k, random_state=42)
            self.labels_ = model.fit_predict(X_scaled)
            self.cluster_centers_ = model.cluster_centers_
        
        return self
    
    def fit_predict(self, X):
        """
        拟合聚类模型并返回聚类标签
        
        参数:
        X: 输入数据
        
        返回:
        labels: 聚类标签
        """
        self.fit(X)
        return self.labels_
    
    def plot_evaluation_graph(self):
        """
        绘制评估图和L方法确定的最佳聚类数
        
        返回:
        plt: matplotlib图形对象
        """
        if self.eval_results is None:
            raise ValueError("必须先调用fit方法")
        
        title = f"{self.method.capitalize()}聚类 + L方法 (使用{self.eval_metric}指标)"
        return visualize_l_method(self.eval_results, self.optimal_k, title=title)
    
    def plot_clusters(self, X, dimensions=[0, 1], cmap='viridis', alpha=0.7, figsize=(10, 8)):
        """
        绘制聚类结果
        
        参数:
        X: 输入数据
        dimensions: 要可视化的维度索引
        cmap: 颜色映射
        alpha: 透明度
        figsize: 图形大小
        
        返回:
        plt: matplotlib图形对象
        """
        if self.labels_ is None:
            raise ValueError("必须先调用fit方法")
        
        plt.figure(figsize=figsize)
        
        # 确保维度索引有效
        if max(dimensions) >= X.shape[1]:
            raise ValueError(f"维度索引{dimensions}超出范围，数据只有{X.shape[1]}个维度")
        
        # 如果维度数量为1，添加一个虚拟维度
        if len(np.unique(dimensions)) == 1:
            X_plot = np.column_stack((X[:, dimensions[0]], np.zeros(X.shape[0])))
            dim_names = [f"维度 {dimensions[0]}", "虚拟维度"]
        else:
            X_plot = X[:, dimensions]
            dim_names = [f"维度 {d}" for d in dimensions]
        
        # 绘制散点图
        scatter = plt.scatter(X_plot[:, 0], X_plot[:, 1], c=self.labels_, cmap=cmap, alpha=alpha)
        
        # 添加聚类中心（如果是K-means）
        if self.method == 'kmeans' and self.cluster_centers_ is not None:
            centers = self.cluster_centers_
            plt.scatter(
                centers[:, dimensions[0]], 
                centers[:, dimensions[1]] if len(dimensions) > 1 else np.zeros(centers.shape[0]),
                marker='X', s=200, c='red', alpha=0.8, label='聚类中心'
            )
        
        plt.colorbar(scatter, label='聚类标签')
        plt.xlabel(dim_names[0])
        plt.ylabel(dim_names[1])
        plt.title(f"{self.method.capitalize()}聚类 (k={self.optimal_k})")
        if self.method == 'kmeans':
            plt.legend()
        plt.grid(True, alpha=0.3)
        
        return plt
    
    def plot_dendrogram(self, X, max_distance=None, figsize=(12, 8), title=None):
        """
        绘制层次聚类的树状图
        
        参数:
        X: 输入数据
        max_distance: 树状图的最大距离，默认为None（自动确定）
        figsize: 图形大小
        title: 图形标题
        
        返回:
        plt: matplotlib图形对象
        """
        if self.method != 'hierarchical':
            raise ValueError("只有层次聚类支持树状图可视化")
        
        plt.figure(figsize=figsize)
        
        # 标准化数据
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
        
        # 计算链接矩阵
        linked = linkage(X_scaled, method=self.linkage, metric=self.metric)
        
        # 绘制树状图
        dendrogram(
            linked,
            truncate_mode='level', 
            p=5,  # 仅显示5个级别
            leaf_rotation=90.,
            leaf_font_size=10.,
            show_contracted=True  # 显示分支中包含的叶子数量
        )
        
        # 在树状图上标记最佳聚类数量
        if max_distance is None:
            # 尝试从链接矩阵推断最佳截断高度
            distances = linked[:, 2]
            sorted_distances = np.sort(distances)
            if len(sorted_distances) >= self.optimal_k:
                # 取第optimal_k-1个最大距离作为截断点
                max_distance = sorted_distances[-(self.optimal_k-1)]
        
        if max_distance is not None:
            plt.axhline(y=max_distance, c='r', linestyle='--', label=f'截断高度 (k={self.optimal_k})')
        
        plt.title(title or f"层次聚类树状图 (最佳聚类数: {self.optimal_k})")
        plt.xlabel('样本索引或分支中的样本数量')
        plt.ylabel('距离')
        plt.legend()
        
        return plt

# 示例：使用自动聚类算法
def demo_auto_cluster():
    # 1. 创建模拟数据集
    n_samples = 1000
    X, y_true = make_blobs(
        n_samples=n_samples, 
        centers=5, 
        cluster_std=[0.8, 1.0, 0.5, 1.2, 0.7],
        random_state=42
    )
    
    print("生成模拟数据集，真实聚类数量：5")
    
    # 2. 使用AutoCluster进行聚类并自动确定聚类数量
    # 2.1 使用层次聚类 + L方法
    print("\n使用层次聚类 + L方法...")
    hc_model = AutoCluster(
        min_clusters=2,
        max_clusters=20,
        method='hierarchical',
        linkage='ward',
        eval_metric='distortion',
        l_method_type='iterative'
    )
    
    hc_labels = hc_model.fit_predict(X)
    print(f"层次聚类确定的最佳聚类数量: {hc_model.optimal_k}")
    
    # 2.2 使用K-means + L方法
    print("\n使用K-means + L方法...")
    km_model = AutoCluster(
        min_clusters=2,
        max_clusters=20,
        method='kmeans',
        eval_metric='distortion',
        l_method_type='iterative'
    )
    
    km_labels = km_model.fit_predict(X)
    print(f"K-means确定的最佳聚类数量: {km_model.optimal_k}")
    
    # 3. 绘制评估图
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    hc_model.plot_evaluation_graph()
    plt.title("层次聚类 + L方法")
    
    plt.subplot(1, 2, 2)
    km_model.plot_evaluation_graph()
    plt.title("K-means + L方法")
    
    plt.tight_layout()
    plt.show()
    
    # 4. 绘制聚类结果
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    scatter1 = plt.scatter(X[:, 0], X[:, 1], c=hc_labels, cmap='viridis', alpha=0.7)
    plt.colorbar(scatter1, label='聚类标签')
    plt.title(f"层次聚类 (k={hc_model.optimal_k})")
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    scatter2 = plt.scatter(X[:, 0], X[:, 1], c=km_labels, cmap='viridis', alpha=0.7)
    plt.colorbar(scatter2, label='聚类标签')
    centers = km_model.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=200, c='red', alpha=0.8, label='聚类中心')
    plt.title(f"K-means聚类 (k={km_model.optimal_k})")
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 5. 绘制层次聚类的树状图
    hc_model.plot_dendrogram(X)
    plt.show()

# 使用不同评估指标的比较
def compare_evaluation_metrics():
    # 创建模拟数据集
    n_samples = 1000
    X, y_true = make_blobs(
        n_samples=n_samples, 
        centers=5, 
        cluster_std=[0.8, 1.0, 0.5, 1.2, 0.7],
        random_state=42
    )
    
    # 评估指标列表
    metrics = ['distortion', 'silhouette', 'calinski_harabasz', 'davies_bouldin']
    
    # 为每个评估指标运行AutoCluster
    results = []
    for metric in metrics:
        model = AutoCluster(
            min_clusters=2,
            max_clusters=20,
            method='kmeans',
            eval_metric=metric,
            l_method_type='iterative'
        )
        
        model.fit(X)
        results.append({
            'metric': metric,
            'optimal_k': model.optimal_k
        })
        
        # 绘制评估图
        plt.figure(figsize=(8, 6))
        model.plot_evaluation_graph()
        plt.title(f"K-means + L方法 (使用{metric}指标)")
        plt.tight_layout()
        plt.show()
    
    # 显示结果比较
    results_df = pd.DataFrame(results)
    print("不同评估指标的比较:")
    print(results_df)
    print("\n真实聚类数量: 5")

if __name__ == "__main__":
    demo_auto_cluster()
    compare_evaluation_metrics()
```
